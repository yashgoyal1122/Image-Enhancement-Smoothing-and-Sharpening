```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def apply_convolution(image, kernel):
    # Apply convolution
    return cv2.filter2D(image, -1, kernel)

def image_smoothing(image, kernel_size):
    # Create a kernel for image smoothing
    kernel = np.ones((kernel_size, kernel_size), np.float32) / (kernel_size * kernel_size)
    return apply_convolution(image, kernel)

def image_sharpening(image):
    # Create a sharpening kernel
    kernel = np.array([[0, -1, 0],
                       [-1, 5, -1],
                       [0, -1, 0]], np.float32)
    return apply_convolution(image, kernel)

# Load the image
image = cv2.imread('/content/Dip - Image sharpning/img2.png')

# Apply image smoothing
smoothed_image = image_smoothing(image, kernel_size=5)

# Apply image sharpening
sharpened_image = image_sharpening(image)

# Display original, smoothed, and sharpened images using matplotlib
plt.figure(figsize=(12, 6))

plt.subplot(1, 3, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title('Smoothed Image')
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title('Sharpened Image')
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()
```
